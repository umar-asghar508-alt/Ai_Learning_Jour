# Lesson 3 : MAtplotlib
Numbers in a table are hard to understand at a glance. But a chart tells the story instantly. Matplotlib is Python's most popular visualization library — it turns your NumPy arrays and Pandas DataFrames into beautiful charts and graphs.

In ML, visualization is used to:

Explore data before training (spot patterns, outliers)
Evaluate models (plot accuracy, loss curves)
Present results (show predictions vs actual values)

**Snippet**

import matplotlib.pyplot as plt → imports the plotting module, plt is the standard shortcut

plt.plot(x, y) → draws a line connecting x and y values

plt.title() → adds a title at the top

plt.xlabel() / plt.ylabel() → labels the axes

plt.show() → displays the plot — always the last line

***Styling***

color='red' → line color

linewidth=2 → thickness of line

linestyle='--' → dashed line ('-' = solid, ':' = dotted)

marker='o' → adds circle at each data point ('s' = square, '^' = triangle)

plt.grid(True) → adds background grid — easier to read values

Multiple plt.plot() calls add multiple lines to same chart

label= → name shown in legend

plt.legend() → displays the legend box showing which line is which

````python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
# Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 5]
# plt.plot(x, y)
# Styling the plot
plt.plot(x,y,color='red',linewidth=3,linestyle='--',marker='o')
plt.title('my first simple plot with styles ')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()

# multiple line plot
x = [1, 2, 3, 4, 5]
math    = [88, 72, 95, 60, 78]
science = [76, 85, 90, 55, 82]
plt.plot(x,math,color='purple',marker='s')
plt.plot(x,science,label='science',color='blue',marker='o')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Student Marks')
plt.legend()
plt.show()

````

### Concept 2: Bar Charts & Histograms

#### Short Theory — When to Use What?

Bar Chart → comparing categories (e.g. scores per student, sales per city)
plt.bar(x, y) → x = categories, y = values

Each category gets one bar

plt.barh() → horizontal bars — useful when category names are long

x and y labels swap compared to vertical bar chart

Histogram → showing distribution of values (e.g. how many students scored between 60-70?)

plt.hist(data, bins=5) → divides data into 5 equal ranges and counts values in each

bins → number of groups/buckets

edgecolor='black' → adds black border to each bar — easier to read


Both are used constantly in ML to understand your data before training.

### Concept 3: Scatter Plots & Subplots
Scatter Plot → shows relationship between two numerical variables. 

Subplots → multiple charts side by side in one figure — used to compare different visualizations at once

plt.scatter(x, y) → plots individual points, no connecting lines

s=100 → size of each point (bigger number = bigger dots)

marker='o' → shape of points

Pattern here → more hours studied = higher score = positive correlation

````python
import matplotlib.pyplot as plt
import numpy as np
#example 
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores   = [40, 50, 55, 65, 70, 80, 85, 95]
plt.scatter(hours_studied,exam_scores,color='red',marker='o',s=100)
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()

#another example 
karachi_math = [88, 95, 78, 91]
karachi_sci  = [76, 90, 82, 88]
lahore_math  = [55, 42, 65, 70]
lahore_sci   = [85, 40, 70, 75]

plt.scatter(karachi_math, karachi_sci, color='blue',  label='Karachi', s=100)
plt.scatter(lahore_math,  lahore_sci,  color='red',   label='Lahore',  s=100)
plt.title('Math vs Science by City')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.legend()
plt.grid(True)
plt.show()
````

### Concept 4: Plotting Directly with Pandas
So far you've been passing lists to Matplotlib manually. But in real ML projects your data is always in a DataFrame. Pandas has a built-in .plot() method that connects directly to Matplotlib — letting you visualize DataFrames in one clean line without manually extracting columns.

#### for histogram
Histogram from DataFrame

df[['Math', 'Science']] → select columns to plot

kind='hist' → histogram

alpha=0.7 → transparency (0=invisible, 1=solid) — lets overlapping

bars show through

Both columns plotted on same histogram automatically

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Student': ['Umar', 'Ali', 'Sara', 'John', 'Ayesha'],
        'Math':    [88, 55, 95, 42, 78],
        'Science': [76, 85, 90, 40, 82]}
df=pd.DataFrame(data)
print(df)
print(df.describe())
print(df.info())
df.plot(x='Student',y='Math',kind='bar',color='red')
plt.xlabel('Student')
plt.ylabel('Math Score')
plt.title('Student vs Math Score')
plt.grid(True)

df.plot(x='Student', y=['Math', 'Science'], kind='bar', figsize=(10, 5))
plt.title('Math vs Science Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
#data 2 with line
data2 = {'Month':    [1, 2, 3, 4, 5, 6],
         'Sales':    [150, 200, 180, 250, 300, 280],
         'Expenses': [100, 120, 110, 160, 180, 170]}

df2 = pd.DataFrame(data2)
# Multiple columns plotted automatically with different colors, no need to mention it 
df2.plot(x='Month',y=['Sales','Expenses'],kind='line',figsize=(12,4),linestyle='--',linewidth=2,marker='o',color=['red','green'],label=['Sales','Expenses']) 
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.grid(True)
# for histogram
df[['Math', 'Science']].plot(kind='hist', bins=5, alpha=0.7, figsize=(8, 5)) # it will overlapped and you will see the difference in original
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
````

## Phase 1 → Lesson 3 → Concept 5: Customizing Plots

A chart that's hard to read is useless. In ML projects you present results to teammates, professors or clients — your charts need to be clear, readable and professional. Customization is what separates a beginner's plot from a professional one.

plt.figure() → must be called BEFORE any plotting , it is like sub plotting , for every graph it should before otherwise it will overwrite it

figsize=(10, 5) → width=10 inches, height=5 inches

dpi=100 → dots per inch = resolution. Higher = sharper image

**Font Sizes & Title Styling**

fontsize=18 → text size

fontweight='bold' → makes title bold

color='darkblue' → title color

plt.xticks(fontsize=11) → controls size of numbers on x axis label line

plt.yticks(fontsize=11) → controls size of numbers on y axis label line

### **Annotation adding**

```
plt.annotate('Highest!',
             xy=(2, 95),
             xytext=(3, 92),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12,
             color='red')
```

plt.annotate() → adds text + arrow pointing to a specific spot

xy=(2, 95) → where the arrow points to (Laiba bar at index 2, height 95)

xytext=(3, 92) → where the text is placed at specific spot

arrowprops=dict(arrowstyle='->') → arrow stylng 

Used in ML to highlight important points on charts

````python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(10,5),dpi=100)
x = [1, 2, 3, 4, 5]
y = [10, 25, 20, 35, 30]

plt.plot(x, y, color='blue', marker='o')
plt.title('Customized Plot')

fig=plt.figure(figsize=(10,5),dpi=100)
plt.plot(x, y, color='blue', marker='o')
plt.title('Sales Report', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Month', fontsize=13)
plt.ylabel('Sales', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

students = ['Umar', 'Ali', 'Laiba', 'John Wick', 'Anmol']
scores   = [88, 55, 95, 42, 78]

plt.figure(figsize=(10, 5))
plt.bar(students, scores, color='steelblue')

# Annotate highest scorer
plt.annotate('Highest!',
             xy=(2, 95),
             xytext=(3, 92),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12,
             color='red')

plt.title('Student Scores')


plt.show()
````

### Important Concept
1. bars = plt.bar(): Yeh har dande (bar) ko aik "Rectangle Object" bana kar save kar leta hai. draw nhi krta blky detail save krleta ha

2. Ek bar ke andar kya hota hai? (The Secrets)
Har ek  (bar) has 3 y "functions" jo humein uski location batate hain:

    a. bar.get_height(): Bar ki unchai (student ka score).

    b. bar.get_x(): Bar jahan se shuru ho rahi hai (Left side).

    c. bar.get_width(): Bar ki chorai (width).

3. -0.4 ka Logic: Bar hamesha center point se thora peeche se shuru   hoti hai taake wo center mein balance lagay.

      a. Logic: Bar jahan se start (get_x), us mein uski adhi (width/2) add kar do taake text bar ke bilkul center mein aa jaye.

      b. Logic: Bar jitni unchi  (height), us se 1 point upar text likho  
      Text: str(score)

      c. Logic: Jo student ka asli number hai, usay text ki shakal mein likh do.

4. plt.text(X, Y, Text) ka Formula:
 X (Center): get_x() + width/2 (Bar ke bilkul darmiyan mein pounchnay ke liye).
 Y (Top): get_height() + 1 (Bar ke sir se thora upar likhne ke liye).
 Text: str(score) (Number ko text bana kar chapne ke liye).



#### **adding label snippet**

   ``` 
   plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 1,
             str(score),
             ha='center',  #horizontal alignment
             fontsize=11)
              ```


````python
print('Important Concept')

students = ['Umar', 'Ali', 'Laiba', 'John Wick', 'Anmol']
scores   = [88, 55, 95, 42, 78]

plt.figure(figsize=(10, 5))
bars = plt.bar(students, scores, color='steelblue') # it is not creating , it makes object rectangle to store info
for bar in bars:
  print(bar) # so here it is x bar start,width , height of bar and angle
bar=zip(bars,scores)
print('\nzipping information in reference \n')
for i in bar:
  print(i)
print('\nzipping information with separate variables\n')
x=zip(bars,scores)
for x,y in x:
  print(f'the refrence of bar {x} and the scores is {y}') # so now after this we are gonna label each bar reference with score
````

````python
import matplotlib.pyplot as plt

students = ['Umar', 'Ali', 'Laiba', 'John Wick', 'Anmol']
scores   = [88, 55, 95, 42, 78]
fig=plt.figure(figsize=(10,5),dpi=100)

bars=plt.bar(students,scores,color='steelblue')
plt.title('Student Scores')
for x,y in zip(bars,scores): # bars refernece==score correspondence
  #adding snippet of text
  plt.text(x.get_x()+x.get_width()/2,
           x.get_height()+1,
           str(y),
            ha='center',  #horizontal alignment
            fontsize=11)
````

### Saving a Plot as an Image
```
plt.figure(figsize=(10, 5))
plt.plot(x, y, color='blue', marker='o')
plt.title('Saved Plot')

plt.savefig('my_plot.png', dpi=150, bbox_inches='tight')
plt.show()

```
plt.savefig('my_plot.png') → saves plot as PNG file

dpi=150 → resolution of saved image

bbox_inches='tight' → removes extra whitespace around plot

Must call savefig() before show() — otherwise saves blank image

Explaination

Why? Because plt.show() clears the figure from memory after displaying it. If you save after show, you're saving a blank empty canvas. So always save first, then show.
