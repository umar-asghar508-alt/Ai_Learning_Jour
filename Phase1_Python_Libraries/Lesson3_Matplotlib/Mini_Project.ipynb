# Mini-Project
Task 1 — Bar Chart:

Plot Math scores per student as a bar chart. Add value labels on top of each bar. Bold title, proper font sizes.

Task 2 — Line Chart:

Plot Math, Science and English scores for all students on one line chart. Add legend, grid, markers.

Task 3 — Scatter Plot:

Plot Math vs Science scores as scatter plot. Annotate the highest Math scorer with 'Top Math!'.

Task 4 — Histogram:

Plot distribution of Average scores with bins=4. Add title, labels, edgecolor.

Task 5 — 2x2 Subplot Dashboard:

Combine all 4 charts above into one 2x2 subplot figure. Title the whole figure 'Student Performance Dashboard' using plt.suptitle().

Task 6 — GroupBy + Plot:

Using df.groupby('City')[['Math','Science','English']].mean(), plot the result as a grouped bar chart using df.plot().

Task 7 — Save:

Save the final dashboard as 'student_dashboard.png' with dpi=150.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Student':  ['Umar', 'Ali', 'Sara', 'John', 'Ayesha', 'Laiba', 'Anmol', 'John Wick'],
    'City':     ['Karachi', 'Lahore', 'Karachi', 'Lahore', 'Karachi', 'Lahore', 'Karachi', 'Lahore'],
    'Math':     [88, 55, 95, 42, 78, 65, 91, 70],
    'Science':  [76, 85, 90, 40, 82, 70, 88, 75],
    'English':  [91, 68, 88, 74, 79, 83, 85, 72]}
# simple plotting of bar
fig=plt.figure(figsize=(10,5),dpi=100)
df = pd.DataFrame(data)
df['Average'] = (df['Math'] + df['Science'] + df['English']) / 3
print(f'\naverage:score: \n{df['Average']}')
df.plot(x='Student',y='Math',kind='bar',label='Maths',color='red')
plt.title('Maths score per student',fontweight='bold')
plt.xlabel('Student', fontsize=13)
plt.ylabel('Math', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
fig=plt.figure(figsize=(10,5),dpi=100)
# head label above the bar
bars=df.plot(x='Student',y='Math',kind='bar',label='Maths',color='red')
for p,y in zip(bars.patches,df['Math']):
    print(p,y)
    #adding snippet of text label
    plt.text(p.get_x()+p.get_width()/2,
           p.get_height()+1,
           str(y),
            ha='center',  #horizontal alignment
            fontsize=11)
plt.tight_layout()
#task 2
df.plot(x='Student',y=['Math','Science','English'],figsize=(10,6),kind='line',linestyle='-',marker='o',grid=True,label=['Math','Science','English'])
plt.title('Math, Science and English scores for all students',fontweight='bold')
plt.legend()
plt.show()
#Task 3
df.plot(x='Math',y='Science',kind='scatter',figsize=(10,6),title='Math vs Science',s=100)
plt.grid(True)
plt.annotate('Highest Point!',
             xy=(df['Math'].max(), df['Science'].max()),
             xytext=(80,90),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12,
             color='red')
#Task4
df[['Average']].plot(kind='hist', bins=4, alpha=0.7, figsize=(8, 5),edgecolor='red') # it will overlapped and you will see the difference in original
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
#Task 5
# fig=plt.figure(figsize=(10,5),dpi=100)
# axis ka mtlb address of box , kis pe kam krna ha 
fig,axs=plt.subplots(2,2,figsize=(15,12))
df.plot(x='Student',y='Math',kind='bar',label='Maths',ax=axs[0,0])
axs[0,0].set_title('Maths score per student',fontweight='bold')
axs[0,0].set_xlabel('Student', fontsize=13)
axs[0,0].set_ylabel('Math', fontsize=13)
# X-axis ke labels bare (14) aur Y-axis ke chote (10)
axs[0,0].tick_params(axis='x', labelsize=8, rotation=45)
axs[0,0].tick_params(axis='y', labelsize=8)
# head label above the bar
bars=df.plot(x='Student',y='Math',kind='bar',label='Maths',color='red',ax=axs[0,0]) #--> work on this box address otherwise it will make another
for p,y in zip(bars.patches,df['Math']):
    # print(p,y)
    #adding snippet of text label
    axs[0,0].text(p.get_x()+p.get_width()/2,
           p.get_height()+1,
           str(y),
            ha='center',  #horizontal alignment
            fontsize=11)
df.plot(x='Student',y=['Math','Science','English'],kind='line',linestyle='-',marker='o',grid=True,label=['Math','Science','English'],ax=axs[0,1])
axs[0,1].set_title('Math, Science and English scores for all students',fontweight='bold')
axs[0,1].legend()
df.plot(x='Math',y='Science',kind='scatter',title='Math vs Science',s=100,ax=axs[1,0])
axs[1,0].grid(True)
axs[1,0].annotate('Highest Point!',
             xy=(df['Math'].max(), df['Science'].max()),
             xytext=(80,90),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12,
             color='red')
df[['Average']].plot(kind='hist', bins=4, alpha=0.7, edgecolor='red',ax=axs[1,1]) # it will overlapped and you will see the difference in original
axs[1,1].set_title('Score Distribution')
axs[1,1].set_xlabel('Score')
axs[1,1].set_ylabel('Frequency')
plt.tight_layout(pad=5.0)
plt.savefig('student_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()

#Task 6
# fig=plt.figure(figsize=(10,5),dpi=100)
df1=df.groupby('City')[['Math','Science','English']].mean()
print('by group by of every city mean of their on subjects \n',df1)
df1.plot(kind='bar',figsize=(10,6),label=['Math','Science','English']) # --> x-axis will come automatically 
# df.plot(x='City',y='df1',kind='bar') ---> wrong 
plt.title('Average Scores by City')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
