#Selecting and Accessing data
**Line by line:**

df['Name'] → single brackets → grabs one column as a Series

df[['Name', 'Score']] → double brackets → grabs multiple columns as a DataFrame

The inner [] is the list of column names, outer [] is the selector

**Pandas has two ways to select rows:**

iloc → thinks in numbers — like NumPy indexing

loc → thinks in labels — uses the index label

Big difference: iloc[0:2] gives rows 0,1 — loc[0:2] gives rows 0,1,2
(end included!)

 **Selecting Rows AND Columns Together**

iloc[0:2, 1:3] → rows 0 to 1, columns 1 to 2 (by position)

loc[0:2, ['Name', 'Score']] → rows 0 to 2, by column name

**Single accessing**

print(df.iloc[1, 2])           # row 1, column 2 → 90
print(df.loc[1, 'Score'])      # row 1, Score column → 90

**Concept 3: Filtering Data**

one list left brakcet means for row and for conditons / second adjacent list right brackets means for column

df['Goals'] > 22 → creates a boolean Series (True/False for each row)

df[...] → keeps only rows where condition is True

Same concept as NumPy boolean masking — just on a DataFrame now

& → AND — both conditions must be True

| → OR — either condition can be True

Each condition wrapped in ( ) — always required in Pandas!

First df[df['Goals'] > 22] → filters rows
Then [['Player', 'Goals']] → selects specific columns
Chained together in one line — very common pattern in real projects

== 'Brazil' → exact match filter

.isin([...]) → checks if value is inside a list — cleaner than

multiple OR conditions

```python
import pandas as pd
data = {'Student': ['Umar', 'Ali', 'Sara', 'John', 'Ayesha', 'Bilal'],
        'Math':    [88, 55, 95, 42, 78, 65],
        'Science': [76, 85, 90, 40, 82, 70],
        'Grade':   ['A', 'C', 'A', 'F', 'B', 'B']}

df = pd.DataFrame(data)
print(df)
print(f"students with scores greater than 70 in maths\n {df[df['Math']>70]['Student']}")
print(f'students with scores greater than 70 in maths & 75 in science\n {df[(df['Math']>70)&(df['Science']>75)]['Student']}')
print(f'{df[df['Grade'].isin(['A','B'])]['Student']}')
print(f'students who failed in maths \n {df[df['Math']<50][['Student','Math']]}')
print(f'student who scored in maths greater than 60 :\n {df[df['Math']>60][['Student','Grade']]}')
```

# Adding, Updating & Deleting Columns
### CONCEPT 4

```python
import pandas as pd
data = {'Student': ['Umar', 'Ali', 'Sara', 'John'],
        'Math':    [88, 55, 95, 42],
        'Science': [76, 85, 90, 40]}
df=pd.DataFrame(data)
print(df)
#adding new col total
df['Total']=df['Math']+df['Science']  # ---->add values row by row (element-wise, just like NumPy)
print(df)
# new col added automatically to the right of data frame
# add passing col
df['Passing']=50
print(df)
#adding result col with conditions
df['Result']=df['Math']>=50 # it will give boolean
print(df)
#update an existing column by maping
df['Result']=df['Result'].map({True:'Pass',False:'Fail'})
print(df)
# updatin an column
df['Math']=df['Math']+5
print(df)
# changing student name john to jhon wick
df.loc[df['Student']=='John','Student']='John Wick'
print(f'method 1  : \n\n{df}')
df.loc[3,'Student']='John Wick'
print(f'method 2  : \n{df}\n')
print(df)
df.drop('Passing',axis=1,inplace=True) # inplace means do change and save in existing dataframe and do not create new one.
print(df)
# df = → must reassign! drop() doesn't modify original unless you add inplace=True
# RENAMING AN COL NAME
df.rename(columns={'Math':'Maths_SCORE'},inplace=True)
print(df)
```

## CONCEPT 5: Handling Missing DAta
### Detecting missing values
None → Python's way of representing missing data

Pandas converts it to NaN (Not a Number) internally

df.isnull() → returns boolean DataFrame — True = missing

df.isnull().sum() → counts missing values per column — most used 

command for checking data quality

###Droping missing values
df.dropna() → removes every row that has at least one missing value

subset=['Math'] → only drops rows where Math is missing — other missing values kept-remove from specified col

Original df unchanged — must reassign

###Filling missing values
df.fillna(0) → replaces every NaN with 0

df['Math'].mean() → calculates mean ignoring NaN values automatically

Filling with mean is the most common strategy for numerical columns in ML

Filling with a label like 'Unknown' is common for text columns


```python
import pandas as pd
import numpy as np

data = {'Student': ['Umar', 'Ali', 'Sara', 'John', 'Ayesha'],
        'Math':    [88, None, 95, 42, 78],
        'Science': [76, 85, None, 40, 82],
        'Grade':   ['A', 'C', 'A', None, 'B']}

df = pd.DataFrame(data)

print(df.isnull(),'\n')        # True where value is missing
print(df.isnull().sum(),'\n')  # count of missing values per column
dro=df.dropna() # drop every row where missing values are
print(dro,'\n')
print(f'before dropping maths rows \n {df}  \n')
dro1=df.dropna(subset=['Math']) # drop row from maths col where missing values are
print(f'after dropping  maths rows\n {dro1} \n')

#FILLING MISSING VALUES
df1=df.fillna(0) # fill missing values with 0
print(f'filling with 0 \n{df1}\n')
df['Math']=df['Math'].fillna(df['Math'].mean()) # fill missing values with mean
print(f'filling with  mean maths col\n{df}\n')
df['Grade']=df['Grade'].fillna('Unknown') # fill missing values with Unknown
print(f'filling with Unknown Grade col \n{df}\n')

```

```python
import pandas as pd
import numpy as np

data = {'Player':  ['Ronaldo', 'Messi', 'Neymar', 'Mbappe', 'Salah'],
        'Goals':   [25, None, 18, 22, None],
        'Assists': [10, 15, None, 18, 9],
        'Country': ['Portugal', 'Argentina', None, 'France', 'Egypt']}

df = pd.DataFrame(data)
print(f'Actual Data \n{df}\n')
print(f'\nno. of missing values of each col\n {df.isnull().sum()}\n')
dro_1=df.dropna()
print(f'dropping all rows having missing values\n{dro_1}\n')
print(f'no. of rows remains after removing : {dro_1.shape[0]}\n')
df['Goals']=df['Goals'].fillna(df['Goals'].mean())
print(f'filling missing values with mean of goals col\n{df}\n')
df['Assists']=df['Assists'].fillna(0)
print(f'filling missing values with 0 in assists col\n{df}\n')
df['Country']=df['Country'].fillna('Unknown')
print(f'filling missing Country with Unknown \n{df}')

```
