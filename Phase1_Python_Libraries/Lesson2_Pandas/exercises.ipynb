Concept 1 — The Two Core Structures

Pandas has two main building blocks:

1. Series — 1D labeled array

pd.Series([...]) → creates a 1D labeled array

Left column = index (labels), Right column = values

dtype: int64 → tells you the data type

```python
import pandas as pd
s=pd.Series([10,20,30,40,50])
print(s)
```

# DataFrame — 2D labeled table
data → a Python dictionary where keys = column names, values = lists

pd.DataFrame(data) → converts dictionary into a table

Left column = index (row numbers), other columns = your data

**important functions**

df.shape → same as NumPy, rows × columns

df.dtypes → shows if column is int, float, string (object) etc.

df.head(n) → shows first n rows — used constantly when exploring new data

df.tail(n) → shows last n rows

/==========================================================
df.info() → most useful function which tells you everything
and it's output of the below program would be like this....

RangeIndex: 5 entries → 5 rows

Non-Null Count: 5 → no missing values in any column

object → text/string data type

int64 → integer data type

memory usage → how much RAM this DataFrame uses

```python
# @title
import pandas as pd
data = {
    'prodcut':['laptop','Phone','Tablet','Watch','Earbuds'],
    'PRice':[1200,800,450,300,150],
    'Stock':[30,50,20,80,100]
}
pd= pd.DataFrame(data) # convert dictionary into table data
print(pd)
print(f'no. of rows & cols : {pd.shape}')
print(pd.dtypes)
#printing first three row
print(pd.head(3))
#printing last two rows
print(pd.tail(2))
#information
print(pd.info())
```

#Selecting and Accessing data
**Line by line:**

df['Name'] → single brackets → grabs one column as a Series

df[['Name', 'Score']] → double brackets → grabs multiple columns as a DataFrame

The inner [] is the list of column names, outer [] is the selector

**Pandas has two ways to select rows:**

iloc → thinks in numbers — like NumPy indexing

loc → thinks in labels — uses the index label

Big difference: iloc[0:2] gives rows 0,1 — loc[0:2] gives rows 0,1,2
(end included!)

 **Selecting Rows AND Columns Together**

iloc[0:2, 1:3] → rows 0 to 1, columns 1 to 2 (by position)

loc[0:2, ['Name', 'Score']] → rows 0 to 2, by column name

**Single accessing**

print(df.iloc[1, 2])           # row 1, column 2 → 90
print(df.loc[1, 'Score'])      # row 1, Score column → 90

**Practise Set 1**

```python
import pandas as pd
data = {'Student': ['Umar', 'Ali', 'Sara', 'John Wick', 'Ayesha'],
        'Math':    [88, 72, 95, 60, 78],
        'Science': [76, 85, 90, 55, 82],
        'English': [91, 68, 88, 74, 79]}

df = pd.DataFrame(data)
print(df)
print(f'maths column only\n {df['Math']}')
print(f'Students and English col\n {df[['Student','English']]}')
print(f'first three rows\n {df.iloc[0:3]}') # with iloc
print(f'1 to 3 rows\n {df.loc[1:3]}') # with loc
print(f'\nsara science score\n {df.loc[2,'Science']}')
```

**Practise set 2 **

```python
import pandas as pd


data = {'Player':  ['Ronaldo', 'Messi', 'Neymar', 'Mbappe', 'Salah'],
        'Goals':   [25, 30, 18, 22, 28],
        'Assists': [10, 15, 12, 18, 9],
        'Rating':  [9.2, 9.5, 8.8, 9.0, 9.1],
        'Country': ['Portugal', 'Argentina', 'Brazil', 'France', 'Egypt']}

df = pd.DataFrame(data)
print(df)
print(f'goals column as series\n {df['Goals']}')
print(f'\n goals,rating and players\n {df[['Goals','Rating','Player']]}')
print(f'\nlast two rows \n{df.iloc[3:5]}')
print(f'\nrow 1 to 3 by loc \n {df.loc[1:3]}')
print(f'\n{df.loc[2,'Player']} rating using iloc :  {df.iloc[2,3]}')
print(f'\n{df.loc[3,'Player']} : {df.loc[3,'Assists']}')
print(f'\n 0 to 2 with player and country \n{df.loc[0:2,['Player','Country']]}')
print(f'\n 2 to 4 rows with first three col \n{df.iloc[2:5,0:3]}')
print(f'Rating of last player {df.loc[df.index[-1],'Player']}: {df.iloc[-1,-2]}') # loc is based on label so it will not work with -index, use df.index[-1]
```

**Concept 3: Filtering Data**

one list left brakcet means for row and for conditons / second adjacent list right brackets means for column

df['Goals'] > 22 → creates a boolean Series (True/False for each row)

df[...] → keeps only rows where condition is True

Same concept as NumPy boolean masking — just on a DataFrame now

& → AND — both conditions must be True

| → OR — either condition can be True

Each condition wrapped in ( ) — always required in Pandas!

First df[df['Goals'] > 22] → filters rows
Then [['Player', 'Goals']] → selects specific columns
Chained together in one line — very common pattern in real projects

== 'Brazil' → exact match filter

.isin([...]) → checks if value is inside a list — cleaner than

multiple OR conditions

```python
import pandas as pd
data = {'Student': ['Umar', 'Ali', 'Sara', 'John', 'Ayesha', 'Bilal'],
        'Math':    [88, 55, 95, 42, 78, 65],
        'Science': [76, 85, 90, 40, 82, 70],
        'Grade':   ['A', 'C', 'A', 'F', 'B', 'B']}

df = pd.DataFrame(data)
print(df)
print(f"students with scores greater than 70 in maths\n {df[df['Math']>70]['Student']}")
print(f'students with scores greater than 70 in maths & 75 in science\n {df[(df['Math']>70)&(df['Science']>75)]['Student']}')
print(f'{df[df['Grade'].isin(['A','B'])]['Student']}')
print(f'students who failed in maths \n {df[df['Math']<50][['Student','Math']]}')
print(f'student who scored in maths greater than 60 :\n {df[df['Math']>60][['Student','Grade']]}')
```
