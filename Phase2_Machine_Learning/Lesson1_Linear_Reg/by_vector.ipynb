```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([92,56,88,70,80,49,65,35,66,67])
y=np.array([98,68,81,80,83,52,66,30,68,73])
plt.plot(x,y,color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
plt.pause(4)
cor=np.corrcoef(x,y)
print(f"the corealatioon beween them is {cor}")
#taking w and b ==0 and n
w=b=0
n=len(x)
learning_rate=0.000001
iteration=3
fcfl=np.zeros(iteration)
#making model
for i in range(iteration):
    for j in range(1):
        py=w*x+b
        print(py)
        print(f"the py value is {py}")
        #making cost funtion MSE one for one w ,b /take mean first/then square
        mean=y-py
        print(f"the value of mean is {mean}")
        me_sq=mean**2
        print(f"the value of mean squre is {me_sq}")
        mse=sum(me_sq)/n
        print(f"the value of mse {mse}")
        #making gradient descendent for one cost function w,b
        #for w
        me_x=-x*mean
        me_x_sum=sum(me_x)
        derrivative_w=2*(me_x_sum)/n
        print(f"the derrivative of w {derrivative_w}")
        #for b
        me_x_neg=-mean
        vsl=sum(me_x_neg)
        derrivative_b=2*(vsl)/n
        print(f"the derrivative of b {derrivative_b}")
        w=w-learning_rate*derrivative_w
        print(f"w {w}")
        b=b-learning_rate*derrivative_b
        print(f"b {b}")
        fcfl[i]=mse
        print(f"cost function {mse} at w {w} and {b}")
finally_pred=w*x+b
plt.scatter(x,y,color='blue')
plt.plot(x,finally_pred,color='green',label='teg line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
plt.show()
   
plt.figure(2)
# not w versus fcfl but iteration vesus fcfl
iteration=np.arange(iteration)
plt.plot(iteration,fcfl,color='purple')
plt.xlabel("Iterations")
plt.ylabel("Cost Function")
plt.title("Cost vs Iterations")
plt.show()

```
