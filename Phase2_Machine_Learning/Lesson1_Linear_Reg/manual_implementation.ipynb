```python

import matplotlib.pyplot as plt
import numpy as np
plt.figure(1)


x=np.array([92,56,88,70,80,49,65,35,66,67])
y=np.array([98,68,81,80,83,52,66,30,68,73])
plt.scatter(x,y,marker='o',color="green")
# plt.plot(x,y,color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
plt.pause(4)
cor=np.corrcoef(x,y)
print(f"the corealatioon beween them is {cor}")
iteration=1000
pyl=np.zeros(len(x))
cfl=np.zeros(iteration)
fcfl=np.zeros(iteration)
# cfl2=np.zeros(len(x))
cost_function=None
n=len(x)

w=b=0

learning_rate=0.000001
wl=np.zeros(iteration)
for i in range(iteration):
    add1=0
    add2=0
    sum_cf=0
    print(f"the value of w and b before prediction loop {w}  {b}")
    for j in range (len(x)):
        #model prediction for every xi each eelemnt
        py=w*x[j]+b
        # ek m,b sy ek cost function sb x ke lia  afte inside fucntion
        mean=y[j]-py
        cf=mean**2
        sum_cf=sum_cf+cf
        # one derivation for only one w and b after one insiode loop
        #"Batch Gradient Descent: Sum all gradients first, update weights once per iteration."
        # for w
        dw=-x[j]*mean
        add1=add1+dw
        db=-mean
        # for b
        add2=add2+db
       
        pyl[j]=py
       
       
    # every elemnyt 
    #  Update OUTSIDE inner loop — after seeing ALL data points
    derivative_w=(2*add1)/n
    derivative_b=(2*add2)/n
    w=w-learning_rate * derivative_w
    b=b-learning_rate * derivative_b
    cost_function=(sum_cf)/n
    print("cost function is at w and b ", w , b ,cost_function)
   
    fcfl[i]=cost_function
    # Why? You first collect the gradient from every single data point, then update once. Updating inside the inner loop means you're changing w and b after every single point — that's unstable and inaccurate. 
    # This is called Batch Gradient Descent — collect all gradients, then update once per iteration.
    wl[i]=w
finally_pred=w*x+b
plt.scatter(x,y,color='blue')
plt.plot(x,finally_pred,color='green',label='teg line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
plt.show()
   
plt.figure(2)
# not w versus fcfl but iteration vesus fcfl
iteration=np.arange(iteration)
plt.plot(iteration,fcfl,color='purple')
plt.xlabel("Iterations")
plt.ylabel("Cost Function")
plt.title("Cost vs Iterations")
plt.show()

```
